## Concat & Merging

```python
import pandas as pd

df1 = pd.DataFrame({
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']},
      index = [0,1,2,3])
```

```python
df2 = pd.DataFrame({
    'A': ['A4','A5','A6','A7'],
    'B': ['B4','B5','B6','B7'],
    'C': ['C4','C5','C6','C7'],
    'D': ['D4','D5','D6','D7']},
     index=[4,5,6,7])
```

```python
df1
```

```python
df2
```

```python
#frames = [df1,df2]
```

```python
result = pd.concat([df1,df2])
result
```

# Join

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'], 
                    'C': ['C0', 'C1', 'C2', 'C3'], 
                    'D': ['D0', 'D1', 'D2', 'D3']}, )


df2 = pd.DataFrame({'E': ['A4', 'A5', 'A6', 'A7'], 
                    'F': ['B4', 'B5', 'B6', 'B7'], 
                    'C': ['C0', 'C3', 'C6', 'C7'], 
                    'G': ['D4', 'D5', 'D6', 'D7']}, )
```

```python
merge_outer = pd.merge(df1,df2,on='C',how='outer')
merge_outer
```

```python
merge_inner = pd.merge(df1,df2,on='C',how='inner')
merge_inner
```

```python
# Keep all rows from df1 (the “left” DataFrame). If C matches in df2, merge; otherwise fill with NaN..
merge_left = pd.merge(df1,df2,on='C',how='left')
merge_left
```

```python
# Keep all rows from df2 (the “right” DataFrame). If C matches in df1, merge; otherwise fill with NaN..
merge_right = pd.merge(df1,df2,on='C',how='right')
merge_right
```

# loc vs iloc

The main distinction between the two methods is:

loc gets rows (and/or columns) with particular labels.

iloc gets rows (and/or columns) at integer locations.



```python
s = pd.Series(list('abcdef'),index=[0,1,2,3,4,5])
s
```

```python
# get the value from the label
s.loc[0]
```

```python
# get the value from the index
s.iloc[-1]
```

```python
# values labels between 0 and 1 (inclusive)
s.loc[0:1]
```

```python
# values index between 0 and 1 (exclusive)
s.iloc[-3:]
```

```python
s.iloc[0:1]
```
